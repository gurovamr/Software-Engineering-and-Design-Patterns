```python
import matplotlib.pyplot as plt

df = bundle.telemetry.copy()

for driver in df["Driver"].dropna().unique():
    sub = df[df["Driver"] == driver]
    plt.plot(sub["Distance"], sub["Speed"], label=driver)

plt.xlabel("Distance")
plt.ylabel("Speed")
plt.title("Speed Trace")
plt.legend()
plt.show()
```

```python
driver = "VER"
sub = bundle.telemetry[bundle.telemetry["Driver"] == driver]

plt.figure(figsize=(12, 4))
plt.plot(sub["Distance"], sub["Throttle"], label="Throttle")
plt.plot(sub["Distance"], sub["Brake"], label="Brake")
plt.xlabel("Distance")
plt.ylabel("Input")
plt.title(f"Throttle / Brake Trace - {driver}")
plt.legend()
plt.show()
```

```python
driver = "VER"
sub = bundle.telemetry[bundle.telemetry["Driver"] == driver]

plt.figure(figsize=(7, 7))
plt.scatter(sub["X"], sub["Y"], c=sub["Speed"], s=5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title(f"Track Map colored by Speed - {driver}")
plt.axis("equal")
plt.show()
```
